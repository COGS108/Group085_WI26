## Research Question

Is food desert status associated with differential access to McDonald’s versus Chipotle restaurants across U.S. census tracts? Specifically, do food desert census tracts have a shorter average distance to McDonald’s locations and a lower likelihood of containing a Chipotle location compared to non-food desert tracts?


## Background and Prior Work

Food deserts refer to areas where residents' access to affordable, healthy food options is restricted or nonexistent due to the absence of grocery stores within convenient traveling distance. According to Food Empowerment Project's description, people’s choices about what to eat are severely limited by the options available to them and what they can afford, and many food deserts contain an overabundance of fast food chains selling cheap “meat” and dairy-based foods that are high in fat, sugar and salt.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Previous studies have identified socio-economic factors that correlates with the presence of food desert. In *Characteristics and Influential Factors of Food Deserts*<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2), the authors find that areas with higher level of poverty are more likely to become food deserts, regardless of urban or rural designation of that area. For other factors, such as vehicle availability and use of public transportations, the association with food desert status varies across very dense urban areas, less dense urban areas, and rural areas. This study has shown that in less urban and rural areas, the percentage of racial minority population affects the food desert status of an area, where food deserts tracts have a greater concentration of minorities. The study has also shown differences across geographical locations within the United States, where low-income populations in some regions are less vulnerable to access problems than low income populations in other regions. However, problems exist in this study. One such problem is that during the regression analysis, the study only examines the food deserts status of the census tracts that are classified as low income. While low-income and food access jointly define food desert status, analyzing the two properties concurrently can distort the relationship between poverty-related variables, such as unemployment rate and poverty rate, and food desert status.

An article published by Addy Yater aimed to examine the difference in grocery accessibility between food deserts and non-food deserts in Indianapolis, Indiana<a name="cite_ref-3"></a><sup>3</sup>. In the article, Yater found that poorer areas contained no grocery stores, while wealthier areas had plenty. A limitation of Yater's study, however, is that it focuses exclusively on regions within Indianapolis, Indiana, which may not be representative of the entire United States. Another problem is that, like the previous study, Yater associates low-income areas with food deserts, which can confound the measurement of other variables within the region. In our study, we aim to compare the accessibility of healthy versus unhealthy foods directly in each region, regardless of whether it is classified as a food desert. While McDonald’s is generally regarded as unhealthy, Chipotle is often considered a healthier alternative within the fast-food industry. Therefore, we chose McDonald’s and Chipotle as the unhealthy and healthy food options, respectively, to compare their accessibility in both food desert and non-food desert areas.

This prediction is based on prior research showing that lower-income and minority communities (areas more likely to be classified as food deserts) tend to have a higher density of traditional fast-food restaurants. ### Normalization and Standardization of our dataset

The food desert dataset contains a large number of variables describing demographic, economic, and food accessibility characteristics of census tracts. These variables are measured on very different scales. For example, population-related variables such as `Pop2010` can reach values above 30,000, while variables such as `num_chipotle` or `num_mcdonalds` range only from 0 to a few units. Because of these large differences in magnitude, some variables may disproportionately influence statistical models that rely on distances or gradients.

```python
food_desert_dataset[[
        'num_mcdonalds',
        'num_chipotle',
        'Pop2010',
        'MedianFamilyIncome',
        'PovertyRate',
        'LILATracts_1And10'
]].describe()
```

Since our dataset contains heavily right-skewed distribution, before we perform standardization, we need to log transform our dataset so that the distributions are normal. 

```python
def log_transform(our_data):
    return np.log1p(our_data)

# Copy food_desert_dataset into data
data = food_desert_dataset[[
        'num_mcdonalds',
        'num_chipotle',
        'Pop2010',
        'MedianFamilyIncome',
        'PovertyRate',
        'LILATracts_1And10'
]]

# Exclude binary column LILATracts_1And10
data_transform = data.dropna(how='any').drop(columns='LILATracts_1And10')

# Apply log transform
data_transform[['Pop2010','MedianFamilyIncome','PovertyRate']] = data_transform[[
    'Pop2010','MedianFamilyIncome','PovertyRate']].apply(log_transform)

print(data.head())
print(data_transform.head())


```

After we log transform the data, we apply StandardScaler to standardize these columns with a mean of zero and standard deviation of 1.

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the log-transformed columns
data_scaled_array = scaler.fit_transform(data_transform)

# Convert back to a dataframe
data_scaled = pd.DataFrame(data_scaled_array,columns=data_transform.columns)

data_scaled['LILATracts_1And10'] = data['LILATracts_1And10']
```

Let's analyze the correlations between each variables, we bring back `LILATracts_1And10` into the dataset to see if there is any correlation between each variables to the census track.

```python
corr_matrix = data_scaled.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Matrix")
plt.savefig('results/correlation_matrix.png')
plt.show()
```

From the correlation matrix, we see that correlation between `num_mcdonalds` after scaling vs `LILATracts_1And10` has a value of 0.02, while the correlation between `num_chipotle`after scaling vs `LILATracts_1And10` has a value of -0.03. Both correlation are weak. 

Let's verify this correlation with OLS, we conduct two OLS regression for `num_mcdonalds` and `num_chipotle` against `LILATracts_1And10`.

```python
X = sm.add_constant(data_scaled[['num_mcdonalds']])
y = data_scaled['LILATracts_1And10']

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
X = sm.add_constant(data_scaled[['num_chipotle']])
y = data_scaled['LILATracts_1And10']

model = sm.OLS(y, X).fit()
print(model.summary())
```

We can see from the OLS regressions that test whether the number of McDonald's or number of Chipotle predicts whether a census tract is a food desert, both models produced R-squared values that are very close to zero, indicating that McDonald's or Chipotle presence explains virtually none of the variation in food desert status. Although the coefficienta were statistically significant(p values less than 0.05), the effect sizes were extremely close to 0, suggesting that the number of McDonald's or Chipotle has little practical impact on whether a tract is classified as a food desert.

## Discussion and Conclusion

In this study, we tested the hypothesis that the number of McDonald’s locations would be positively correlated with food desert status, as measured by LILA, and that the number of Chipotle locations would be negatively correlated with LILA. However, our analyses revealed that there appears to be no meaningful correlation between the number of McDonald’s locations and LILA scores, nor between the number of Chipotle locations and LILA. This indicates that the presence of these chain restaurants alone does not serve as a reliable indicator of whether an area is a food desert, and therefore our hypothesis was not supported.

This work builds on previous research on food deserts, which emphasized that access to affordable healthy food is shaped by multiple social-economic and geographic factors. Prior study highlighted the influence of poverty, vehicle availability, racial composition, and regional variation on food desert status. Our study contributes to this body of work by examining fast-food accessibility directly, comparing generally unhealthy and relatively healthier options across census tracts without relying solely on income or urban/rural designation.

There are limitations to our study. Our analysis is cross-sectional, capturing only a single time point, which prevents us from assessing temporal dynamics or causal relationships. The dataset may also overlook smaller food retailers, local markets, or non-chain fast food options that contribute to food access, potentially biasing the results. A more granular or longitudinal dataset could provide deeper insights into how food access evolves over time and whether restaurant presence has indirect effects not captured here.

Considering all the results together, it seems that food desert status is influenced by factors beyond the placement of major chain restaurants. Areas lacking McDonald’s or Chipotle are not necessarily food deserts, and conversely, some food deserts may have one or both of these chains present. This finding highlights the complexity of food accessibility, which is likely shaped by local grocery availability, socioeconomic conditions, transportation infrastructure, and regional policies, rather than simply the presence of fast food chains.
